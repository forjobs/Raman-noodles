```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import jcamp
import peakutils
```

```python
# import CO2 .jdx
CO2_data = jcamp.JCAMP_reader('124-38-9-IR.jdx')
```

```python
y_abs = 1 - CO2_data['y']
CO2_data['yunits'] = 'ABSORBANCE'
CO2_data['y'] = y_abs
CO2_data
```

```python
indexes = peakutils.indexes(CO2_data['y'], thres=0.5, min_dist=30)

plt.figure(figsize=(10,6))
plt.plot(CO2_data['x'], CO2_data['y'])
for i in range(len(indexes)):
    plt.axvline(x=indexes[i], color='orange')
```
